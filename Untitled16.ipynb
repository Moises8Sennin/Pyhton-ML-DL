```python
class Model(nn.Module):

    def __init__(self, inp=9, h1=64, h2=32,h3=32,h4=16,outp=3):
        super().__init__()
        self.fc1 = nn.Linear(inp, h1)
        self.fc2 = nn.Linear(h1,h2)
        self.fc3 = nn.Linear(h2,h3)
        self.fc4 = nn.Linear(h3,h4)
        self.out = nn.Linear(h4,outp)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = self.out(x)

        return x
        
        
        
    
```

```python
torch.manual_seed(41)
model = Model()
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_sc = x_sc.values
y = y.values
```

```python
X_tr, X_ts, Y_tr, Y_ts = train_test_split(x_sc,y, test_size=0.2, random_state=41)
```

```python
type(Y_ts)
```

```python
X_tr = torch.FloatTensor(X_tr)
X_ts = torch.FloatTensor(X_ts)
```

```python
Y_tr = torch.LongTensor(Y_tr)
Y_ts = torch.LongTensor(Y_ts)
```

```python
Y_tr.shape
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=0.001)
```

```python
epochs = 2000
losses = []
train_accuracies = []
val_accuracies = []
for i in range(epochs):
    y_pred = model.forward(X_tr)

    loss = criterion(y_pred, Y_tr)
    losses.append(loss.detach().numpy())

    _, predict = torch.max(y_pred, 1)
    train_accur = (predict == Y_tr).float().mean().item()
    train_accuracies.append(train_accur)

    with torch.no_grad():
        val_out = model.forward(X_ts)
        _, val_pred = torch.max(val_out,1)
        val_accur = (val_pred== Y_ts).float().mean().item()
        val_accuracies.append(val_accur)

    if i % 10==0:
        print(f'Epoch: {i} and loss : {loss} train_accuracy : {train_accur} validation_accuracy : {val_accur}') 
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
len(val_accuracies)
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.plot(range(epochs), train_accuracies, label='Train')
plt.plot(range(epochs), val_accuracies, label='Validation')
plt.title("learning curve")
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```
